```python
import numpy as np
import pandas as pd
from pyecharts import options as opts
from pyecharts.charts import Bar
import folium
from folium import plugins
```

```python
data = pd.read_csv('crawler_output.csv',encoding='utf-8')
```

```python
type2 = data['type2'].value_counts()
```

```python
labels = list(type2.index)
value1 = list(type2.values)
```

```python
bar = (
  Bar()
  .add_xaxis(labels) 
  .add_yaxis("类别", value1) 
  .set_global_opts(title_opts=opts.TitleOpts(title="休闲活动类别排行")) #设置标题
)
bar.render('entertainment.html')
```
